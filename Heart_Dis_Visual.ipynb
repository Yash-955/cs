```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_data = pd.read_csv("C:/Users/lenovo/Downloads/Heart.csv")
```

```python
# 1. Age Distribution by Heart Disease Status
plt.figure(figsize=(10, 6))
sns.histplot(data=heart_data, x='age', hue='target', kde=True, bins=20)
plt.title('Age Distribution by Heart Disease Status')
plt.show()
```

```python
# 2. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heart_data.corr(), annot=True, cmap='RdYlGn')
plt.title('Correlation Matrix of Heart Disease Data')
plt.show()
```

```python
# 3. Chest Pain Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=heart_data, x='cp', hue='target', palette='Set2')
plt.title('Chest Pain Type Distribution by Heart Disease Status')
plt.show()
```

```python
# 4. Boxplot for Cholesterol Levels
plt.figure(figsize=(8, 5))
sns.boxplot(x='target', y='chol', data=heart_data, palette='pastel')
plt.title('Cholesterol Levels by Heart Disease Status')
plt.show()
```

```python
# 5. Scatter Plot of Max Heart Rate vs Age
plt.figure(figsize=(8, 5))
sns.scatterplot(data=heart_data, x='age', y='thalach', hue='target', style='sex', palette='cool')
plt.title('Max Heart Rate vs Age by Disease Status and Gender')
plt.show()
```

```python
# 8. Bar Plot for Fasting Blood Sugar
plt.figure(figsize=(6, 4))
sns.countplot(data=heart_data, x='fbs', hue='target', palette='viridis')
plt.title('Fasting Blood Sugar by Heart Disease Status')
plt.show()
```

```python
# 9. Scatter Plot: Oldpeak vs. Age
plt.figure(figsize=(8, 5))
sns.scatterplot(data=heart_data, x='age', y='oldpeak', hue='target', palette='coolwarm')
plt.title('ST Depression (Oldpeak) vs Age')
plt.show()
```
