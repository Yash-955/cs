```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/lenovo/Documents/Facebook.csv') 
```

```python
# Set up the figure
plt.figure(figsize=(10, 6))

# Plotting distribution for likes
sns.histplot(df['like'], kde=True, color='blue', label='Likes')
sns.histplot(df['share'], kde=True, color='green', label='Shares')
sns.histplot(df['comment'], kde=True, color='red', label='Comments')

# Title and labels
plt.title('Distribution of Facebook Metrics')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Scatter plot to see the relationship between likes and shares
sns.scatterplot(x='like', y='share', data=df, hue='Type', palette='Set1')
plt.title('Likes vs Shares')
plt.xlabel('Likes')
plt.ylabel('Shares')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Boxplot for likes, shares, and comments
sns.boxplot(data=df[['like', 'share', 'comment']])
plt.title('Boxplot of Facebook Metrics')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = df[['like', 'share', 'comment']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=1, linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_monthly = df.groupby('Post Month')[['like', 'share']].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Post Month', y='like', data=df_monthly, label='Likes')
sns.lineplot(x='Post Month', y='share', data=df_monthly, label='Shares')

plt.title('Facebook Metrics Over Post Month')
plt.xlabel('Post Month')
plt.ylabel('Count')
plt.legend()
plt.show()
```
